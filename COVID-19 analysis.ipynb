```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
from matplotlib.animation import FuncAnimation
from matplotlib import style
```

```python
total_deaths=pd.read_csv('total_deaths.csv',parse_dates=['date'])
total_deaths.head()
```

```python
total_deaths.set_index('date',inplace=True)
```

```python
total_cases=pd.read_csv('total_cases.csv',parse_dates=['date'])
total_cases.head()
```

```python
total_cases.set_index('date',inplace=True)
total_cases.head()
```

```python
total_cases.shape
```

```python
total_cases['World'].max()
```

```python
total_cases.tail()
```

```python
print(total_cases.index[-1])
```

```python
plt.figure(figsize=(13,12))
plt.title('COVID-19 Reports worldwide')
plt.ylabel('Number of COVID-19 cases')
plt.xlabel('Date')
plt.plot(total_cases.index,total_cases['World'],'r',label='Reported Corona infection cases')
plt.plot(total_deaths.index,total_deaths['World'],'b',label='Reported Corona Deaths')
plt.grid(True,color='g')
plt.legend()
plt.style.use('ggplot')
plt.savefig('covid-19.jpeg')
plt.show()
```
